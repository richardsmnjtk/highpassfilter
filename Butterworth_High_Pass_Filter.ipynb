```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt
```

```python
from IPython.display import Audio
from scipy.io import wavfile
```

```python
fs = 8000
x = np.arange(fs)
cutoff = 277
nyq = 0.5 * fs
order = 2
```

```python
#Signal Awal
a = np.sin(586*2*np.pi*(x/fs))
b = np.cos(863*2*np.pi*(x/fs))
data = a+b
plt.plot(data)
plt.xlim(0, 100)
```

```python
Audio(data, rate = fs)
```

```python
def filter_highpass(data, cutoff, fs, order):
    normal_cutoff = cutoff/nyq
    b, a = butter(order, normal_cutoff, btype='high')
    y = filtfilt(b, a, data)
    return y
```

```python
y = filter_highpass(data, cutoff, fs, order)
plt.plot(y)
plt.xlim(0, 100)
```

```python
Audio(y, rate = fs)
```
